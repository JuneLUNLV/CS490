```python
def get_urls_from_playlist(url):
    options = webdriver.ChromeOptions()
    options.add_argument('headless')
    driver = webdriver.Chrome(options=options)
    driver.get(url)
    urlArray = []
    xpath = ("//ytd-playlist-video-renderer/div[@id='content']/a")
    linkElementArray = driver.find_elements_by_xpath(xpath)
    for ele in linkElementArray:
        urlArray.append(ele.get_attribute('href'))
    return urlArray
```

```python
def printProgressBar (iteration, total, prefix = '', suffix = '', decimals = 1, length = 100, fill = 'â–ˆ'):
    """
    Call in a loop to create terminal progress bar
    @params:
        iteration   - Required  : current iteration (Int)
        total       - Required  : total iterations (Int)
        prefix      - Optional  : prefix string (Str)
        suffix      - Optional  : suffix string (Str)
        decimals    - Optional  : positive number of decimals in percent complete (Int)
        length      - Optional  : character length of bar (Int)
        fill        - Optional  : bar fill character (Str)
    """
    percent = ("{0:." + str(decimals) + "f}").format(100 * (iteration / float(total)))
    filledLength = int(length * iteration // total)
    bar = '\r' + fill * filledLength + '-' * (length - filledLength)
    tempStr = '{} |{}| {}% {}'.format(prefix, bar, percent, suffix)
    print(tempStr,end='\r')
    # Print New Line on Complete
    if iteration == total: 
        print('',end='')
```
